```python
import pandas as pd
from sklearn.naive_bayes import MultinomialNB, BernoulliNB
from sklearn.svm import LinearSVC
from sklearn.neural_network import multilayer_perceptron
from sklearn.feature_extraction.text import TfidfVectorizer,CountVectorizer,TfidfTransformer
from sklearn.cross_validation import train_test_split
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.naive_bayes import MultinomialNB

# from sklearn.cross_validation import cross_val_score, cross_val_predict
from sklearn import metrics
# Perform 5-fold cross validation
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
```

```python
df = pd.read_csv('data sentiment/dataModelClean.csv', encoding='utf-8')
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vect = CountVectorizer()
X_train_counts = count_vect.fit_transform(df.text)
#print X_train_counts.shape
#count_vect.vocabulary_
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
tfidf_transformer = TfidfTransformer()
X_train_tfidf = tfidf_transformer.fit_transform(X_train_counts)
print X_train_tfidf.shape
print(tfidf_transformer.fit_transform(count_vect.fit_transform(df.text)).toarray())
```

```python
nb = MultinomialNB()
```

```python
# from sklearn.model_selection import KFold
df_actual = df['actual']
data_X = X_train_tfidf
kf = KFold(n_splits=10)
for train, test in kf.split(data_X):
    #print("%s %s\n " % (train, test))
    X = X_train_tfidf
    y = df_actual
    print train
    print test
    X_train1, X_test1, y_train1, y_test1 = X[train], X[test], y[train], y[test]
    print("Size of training set: {} size of test set: {}\n".format(X_train1.shape, X_test1.shape))

scores = cross_val_score(nb, X_train_tfidf, y, cv=10)
print "Cross-validated scores :", scores
    
```

```python
performance_model = []

from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
import pickle
i = 0
kf = KFold(n_splits=10)
for train, test in kf.split(data_X):
    X = X_train_tfidf
    y = df_actual
    X_train, X_test, y_train, y_test = X[train], X[test], y[train], y[test]
    model_combine_feature = nb.fit(X_train,y_train)
    print"classification_report", classification_report(model_combine_feature.predict(X_test),y_test)
    report = classification_report(model_combine_feature.predict(X_test),y_test)
    report = report.split()
    performance_model.append([report[17],report[18],report[19]])
    with open('model_combine '+ str(i)+'.pkl ', 'wb') as f:
        pickle.dump(model_combine_feature, f)
        print('model combine feature',i,' saved in', './model.pkl',i)
        i+= 1
```

```python
a = pd.DataFrame(performance_model)
a.columns = ['precision','recall','f1-score']
a
```
